```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Baixar as biliotecas do Python
```

2) **IMPORTANDO A BASE DE DADOS

```python
base = pd.read_csv('Consumo_cerveja.csv', sep=';')
# Importando a nossa base de dados

base.head()
# Vendo a base de dados

base.info()
# int64 = contém números inteiros
# float64 = contém números decimais
# object = contém texto

sns.pairplot(base)
# Visualizar correlação dos dados

base = base.select_dtypes(include=['number'])
sns.heatmap(base.corr())
# Visualizar correlação dos dados
# Retirei o object que tinha
```
